```python
base_site = "https://en.wikipedia.org/wiki/Music"
response = requests.get(base_site)
response.status_code
```

```python
html = response.content
soup = BeautifulSoup(html, "html.parser")
```

```python
div_notes = soup.findAll("div", {'role':'note'})
div_notes
```

```python
div_notes[0].find('a')
```

```python
div_links = [l.find('a') for l in div_notes]
len(div_links)
```

```python
div_notes[7]
```

```python
all_links = []
for i in div_notes:
    q = i.findAll('a')
    
    all_links.extend(q)
```

```python
len(all_links)
```

```python
all_links
```

```python
note_url = [urljoin(base_site,i.get('href')) for i in all_links]
note_url
```

```python
par_text = []
count = 0
for i in note_url:
    note_response = requests.get(i)
    if note_response.status_code == 200:
        print("URL #{0}: {1}".format(count+1,i))
    else:
        print("Error {0}: URL #{1}:{2} is invalid, hence skipping".format(note_response.status_code,count,i))
        count += 1
        continue
        
    note_html = note_response.content 
    note_soup = BeautifulSoup(note_html,'lxml')
    note_para = note_soup.findAll('p')
    note_para_texts = [p.text for p in note_para]
    par_text.append(note_para_texts)
    count += 1
```

```python
par_text
```

```python
par_join = "".join(par_text[0])
par_join
```

```python
par_join = ["".join(para) for para in par_text]
print(par_join[2])
```

```python
url_and_para = dict(zip(note_url,par_join))
print(url_and_para['https://en.wikipedia.org/wiki/Timbre'])
```
