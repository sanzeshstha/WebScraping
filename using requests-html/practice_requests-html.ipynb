# Exploring the requests-html capabilities

## Initial Setup

```python
from requests_html import HTMLSession
```

```python
session = HTMLSession()
response = session.get("https://en.wikipedia.org/wiki/Association_football")
response.status_code
```

```python
response.html
```

## Links

```python
url = response.html.links     #relative links
len(url)
```

```python
full_links = response.html.absolute_links    #full path links
len(full_links)
```

```python
type(full_links)
```

## Searching for Elements
